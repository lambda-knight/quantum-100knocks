## 解説
一般に1量子ビットの状態は
$$
|\psi\rangle = \alpha|0\rangle + \beta|1\rangle
$$
と表現される。また、規格化条件は以下の通りにあらわされる。
$$
|\alpha|^2 + |\beta|^2 = 1　
$$

ブロッホ球とは量子の状態を分りやすく表現するために球の表面に表示するためのものであり、ブロッホ球の式は以下の通りとなる。

$$
{|\psi〉 = cos(\theta / 2)|0〉 + e^{i\phi}sin(\theta / 2)|1〉\\
ただし、0≦\theta≦\pi、0≦\phi≦2\pi
}
$$


求めるブロッホ球の位置は$\theta=\frac{\pi}{4},\phi=0$となり、$|+\rangle$である。
$|0\rangle$にアダマールゲートを作用させると$|+\rangle$となり、上記のブロッホ球の位置に移動する。
$$
H|0\rangle=\frac{|0\rangle+|1\rangle}{\sqrt{2}}=|+\rangle
$$
また、アダマールゲートはX基底、Z基底では$\frac{\pi}{4}$の回転になるので、$|0\rangle$にアダマールゲートを適用すると上記のようになる。
